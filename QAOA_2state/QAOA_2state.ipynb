```python
import numpy as np

from qutip import * # for handling quantum mechanical operators, vectors, ...
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def U_H(H, angle):
    exponential = (-1j*angle*H)
    return exponential.expm()
```

```python
iterations = 8
gamma = np.linspace(0, 2*np.pi, iterations)
beta = np.linspace(0, np.pi, iterations)

s = (qutip.basis(2, 0) + 1*qutip.basis(2, 1)).unit() # initial state - eigen state of sigma_x
```

```python
font_size = 20
figure, ax = plt.subplots(2,3,figsize=(15,7.5), constrained_layout=True, subplot_kw=dict(projection='3d'))

first_step_state = []

for gammas in gamma:
    first_step_state.append(U_H(sigmaz(), gammas) * s)

second_step_state = []
gamma_fixed = np.pi / 5
for betas in beta:
    second_step_state.append(U_H(sigmax(),betas) * U_H(sigmaz(),gamma_fixed) * s)
    
b_initial = qutip.Bloch(fig=figure, axes=ax[0,0])
b_initial.add_states([s])
b_initial.vector_color = ['r']
b_initial.render(fig=figure, axes=ax[0,0])

b_first = qutip.Bloch(fig=figure, axes=ax[0,1])
b_first.add_states(first_step_state)
b_first.vector_color = ['b']
b_first.render(fig=figure, axes=ax[0,1])

b_second = qutip.Bloch(fig=figure, axes=ax[0,2])
b_second.add_states(second_step_state)#
b_second.vector_color = ['g']
b_second.render(fig=figure, axes=ax[0,2])

gamma_fixed = np.pi / 4
beta_fixed = np.pi / 4

first_step_state = U_H(sigmaz(), gamma_fixed) * s
second_step_state = U_H(sigmax(),beta_fixed) * U_H(sigmaz(),gamma_fixed) * s

b_initial = qutip.Bloch(fig=figure, axes=ax[1,0])
b_initial.add_states([s])
b_initial.vector_color = ['r']
b_initial.render(fig=figure, axes=ax[1,0])

b_first = qutip.Bloch(fig=figure, axes=ax[1,1])
b_first.add_states(first_step_state)
b_first.vector_color = ['b']
b_first.render(fig=figure, axes=ax[1,1])

b_second = qutip.Bloch(fig=figure, axes=ax[1,2])
b_second.add_states(second_step_state)
b_second.vector_color = ['g']
b_second.render(fig=figure, axes=ax[1,2])

ax[1,0].set_title(r"$|s\rangle$", fontsize=font_size)
ax[1,1].set_title(r"$\hat{U} (\hat{H}_P,\gamma = \frac{\pi}{4}) |s\rangle$", fontsize=font_size)
ax[1,2].set_title(r"$\hat{U} (\hat{H}_D,\beta = \frac{\pi}{4}) \hat{U} (\hat{H}_P,\gamma = \frac{\pi}{4}) |s\rangle$", fontsize=font_size)

ax[0,0].set_title(r"$|s\rangle$", fontsize=font_size)
ax[0,1].set_title(r"$\hat{U} (\hat{H}_P,\gamma) |s\rangle$", fontsize=font_size)
ax[0,2].set_title(r"$\hat{U} (\hat{H}_D,\beta) \hat{U} (\hat{H}_P,\gamma = \frac{\pi}{5}) |s\rangle$", fontsize=font_size)

plt.savefig("images/evolution_bloch_sphere_general_and_maximum.png", dpi=200, bbox_inches='tight')
```
